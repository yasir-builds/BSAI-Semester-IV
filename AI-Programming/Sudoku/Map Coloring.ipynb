### 📝 Task Overview
The objective of this lab is to solve the **Map Coloring Problem** using **Constraint Satisfaction (CSP)** logic. The goal is to assign a color to each of the 7 regions such that no two adjacent regions share the same color.

### ⚙️ Constraints
* **Regions:** WA, NT, SA, Q, NSW, V, T
* **Colors:** Red, Green, Blue
* **Adjacency Rule:** Neighboring regions (e.g., WA and NT) must have distinct color assignments.

### 🧠 Logic Flow
1.  **Validation:** The `is_valid` function checks if a color assignment conflicts with existing neighbor data.
2.  **Backtracking:** The `solve_map_coloring` function recursively assigns colors and "backtracks" if a conflict is hit.
3.  **Completion:** The process terminates once all regions are successfully assigned a valid color.

---

```python
regions = ['WA', 'NT', 'SA', 'Q', 'NSW', 'V', 'T']
colors = ['Red', 'Green', 'Blue']
neighbors = {
    'WA': ['NT', 'SA'],
    'NT': ['WA', 'SA', 'Q'],
    'SA': ['WA', 'NT', 'Q', 'NSW', 'V'],
    'Q': ['NT', 'SA', 'NSW'],
    'NSW': ['Q', 'SA', 'V'],
    'V': ['SA', 'NSW'],
    'T': []
}

def is_valid(region, color, assignment):
    for neighbor in neighbors[region]:
        if neighbor in assignment and assignment[neighbor] == color:
            return False
    return True

def solve_map_coloring(assignment):
    if len(assignment) == len(regions):
        return assignment

    unassigned = [r for r in regions if r not in assignment]
    current = unassigned[0]

    for color in colors:
        if is_valid(current, color, assignment):
            assignment[current] = color
            result = solve_map_coloring(assignment)
            if result is not None:
                return result
            assignment.pop(current)
            
    return None

solution = solve_map_coloring({})

if solution:
    print("Map coloring Solution:")
    for region, color in solution.items():
        print(f"{region} -> {color}")
```
