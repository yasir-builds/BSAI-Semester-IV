## Lab 6: Alpha-Beta Pruning
**Objective:** Modify the standard Minimax algorithm to include Alpha-Beta pruning to optimize game tree evaluation by eliminating unnecessary branches.

### Lab Details:
* **CLO:** CLO2
* **PLO:** PLO2
* **Taxonomy Level:** P3
* **Marks:** 4

---

```python
import math

def minimax(depth, nodeIndex, maximizingPlayer, values, alpha, beta):
    # Base case: leaf node reached at depth 3
    if depth == 3:
        return values[nodeIndex]

    if maximizingPlayer:
        best = -math.inf
        for i in range(0, 2):
            val = minimax(depth + 1, nodeIndex * 2 + i, False, values, alpha, beta)
            best = max(best, val)
            alpha = max(alpha, best)

            # Alpha-Beta Pruning condition
            if beta <= alpha:
                break
        return best

    else:
        best = math.inf
        for i in range(0, 2):
            val = minimax(depth + 1, nodeIndex * 2 + i, True, values, alpha, beta)
            best = min(best, val)
            beta = min(beta, best)

            # Alpha-Beta Pruning condition
            if beta <= alpha:
                break
        return best

# Driver Code
if __name__ == "__main__":
    # Example leaf node values
    leaf_values = [3, 5, 6, 9, 1, 2, 0, -1]
    
    # Initial call: depth 0, root index 0, maximizing player, alpha -inf, beta +inf
    optimal_val = minimax(0, 0, True, leaf_values, -math.inf, math.inf)
    
    print("The optimal value is:", optimal_val)
```
